```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv("FDI_in_India (1).csv")
data = dataset.copy()
```

```python
for i in range(len(data)):
    if data.loc[i].value_counts().index[0] == 0:
        print(data.loc[i][0])
```

```python
for i in range(len(data)):
    if data.loc[i].value_counts().index[0] == 0:
        # print(data.loc[i])
        data.drop([i],axis=0,inplace=True)
        # print(data.loc[i].value_counts().index[0])
```

```python
#Performance of each sector recieving significant FDI
sectors = data['Sector'].unique()

# Set the figure size
fig, axes = plt.subplots(nrows=len(sectors), ncols=1, figsize=(10, 6*len(sectors)))

# Iterate over the sectors and plot a line chart for each one
for i, sector in enumerate(sectors):
    sector_data = data[data['Sector'] == sector]
    axes[i].plot(sector_data.columns[1:], sector_data.iloc[0, 1:])
    axes[i].set_title(sector)
    axes[i].set_xlabel('Year')
    axes[i].set_ylabel('Performance')
    axes[i].set_xticklabels(sector_data.columns[1:], rotation=90)

# Adjust the space between subplots and show the chart
plt.subplots_adjust(hspace=0.5)
plt.show()
```
